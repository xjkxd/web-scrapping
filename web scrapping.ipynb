                        EXTRACING IMAGE INFORMATION IN WEB PAGE IN PYTHON

```python
import requests
from bs4 import BeautifulSoup
```

```python
web = requests.get("https://www.tutorialsfreak.com/")
```

```python
soup =  BeautifulSoup(web.content,"html.parser")
```

```python
# print(soup.prettify())
```

```python
img = soup.find_all("img")
```

```python
img
```

```python
for i in img:
    print(i.get("src"))
    print(type(i.get("src")))
```

```python
for i in img:
    print(i.get("alt"))
```
